```python
import pulp as plp
solver = plp.GLPK_CMD()
# Problème de minimisation
model = plp.LpProblem('Minimisation_coût_transport', plp.LpMinimize)

#Données 
cout =  {
    (1,1):2,
    (1,2):3,
    (1,3):1,
    (1,4):4,
    (2,1):5,
    (2,2):2,
    (2,3):3,
    (2,4):2,
    (3,1):3,
    (3,2):4,
    (3,3):2,
    (3,4):1
}

offres = [1, 2, 3]
demande = [1, 2, 3, 4]

capacite = {1:120, 
            2:80, 
            3:100}
demandes_cap = {1:70, 
                2:90, 
                3:60, 
                4:80}

# Variables de décisions
x  = plp.LpVariable.dicts('x', cout.keys(), lowBound=0, cat = 'continuous')

# Fonction objectif
model += plp.lpSum([cout[i, j]*x[i, j] for i in offres  for j in demande])

# Contraintes 
for i in offres:
    model += plp.lpSum([x[i, j] for j in demande]) == capacite[i]

for j in demande:
    model += plp.lpSum([x[i, j] for i in offres]) == demandes_cap[j]


# Résolution 
model.solve(solver)

# Résultats
print("Status : ", plp.LpStatus[model.status])
for (i, j), var in x.items():
    if(var.value() > 0):    
        print(f"x[{i}, {j}] = {var.value()}") 
    
print("Coût Total : ", plp.value(model.objective))
```
